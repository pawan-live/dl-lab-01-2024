```python
# Generate random data
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.uniform(low=-3, high=3, size=100000)
normal_data = np.random.normal(loc=0, scale=1, size=100000)

# Plot histograms for each distribution
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

# Set plot view and labels
plt.axis([-2.5, 6, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")

# Show the plot
plt.show()

plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue",
label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Create the figure and axes object
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(projection="3d")

# Create the meshgrid
num1 = np.arange(-5, 5, 0.1)
num2 = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(num1, num2)

# Plot the surface
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)

# Set the axis labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set the title
ax.set_title("3D Plot of Z = X**2 + Y**2")

# Display the plot
plt.show()
```

```python

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')


# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected = df[features]

# Calculate Pearson correlation coefficient
pearson_corr = df_selected.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = df_selected.corr(method='spearman')

# Visualize using heatmaps
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Rank Correlation Heatmap")
plt.show()
```
